```python

"""" 
    DocString:
        
    This game is based on the Harry Potter movie series. It consists of 3 
    stages that If you have watched the Harry Potter series, you will have an
    easier time in the game, especially in the last two stages. 
    
    First Stage: You will substitute Cedric in the tournament.
    
    Second Stage: Answer 2 questions to gain Harryâ€™s trust.
    
    Third Stage: Help Harry defeat Voldemort
    
    Known Bugs:
       1. On stage 2, the user has to type exactly what the answer is, or the
    number before the answer, or the code will break.
        2. On stage 2, when I included the function "question_list.remove(question)"
    the second question asked sometimes mixed up with the answer. An example is
    question 2 having the answers from question 4. Because of this, I decided
    to keep that function out. Like this at least the questions can match the 
    answers."
"""


from sys import exit
import random

def start():
    # introduce the scenario for the user
    
    print("""
        You just arrived in Hogwarts for your first year in the greatest
        school of witchcraft and wizardry! You are very excited to start learning 
        magic to become an outstanding witch/wizard! A very tall man
        approaches you and engages in conversation!""")
    
    print(f"""\n {'-' * 80}""")
    
    # user gives out his name so that he can feel like he/she is in the game
    
    # fix this blank thing
    username = input(prompt = """ 
        "Hello There! My name is Hagrid and you look
        lost! What is your name?"
    
            <Enter your name> """)

    print(f"""\n
        "It's nice to meet you {username.capitalize()}! I assume you 
        are a first year here at Hogwarts! Here is the sorting hat 
        to decide which house you will be part of!""")
    
    house_list = ['Gryffindor',
                    'Hufflepuff',
                     'Slytherin',
                     'Ravenclaw']
    
    # randomly assigns user a house in Hogwarts
    
    house = random.choice(house_list) 
    input("""\n
        Press 'Enter' so the sorting hat can send you to your house!
    """)
    print(house)
    print(" ")
        
    if house == house_list[0]:
        print(""" 
        "Welcome to Gryffindor! Gryffindor values courage, bravery, 
        nerve, and chivalry!" """)
            
            
    elif house == house_list[1]:
        print(""" 
        "Welcome to Hufflepuff! Hufflepuff values hard work, 
        patience, justice, and loyalty!"" """)
            
            
    elif house == house_list[2]:
        print(""" 
        "Welcome to Slytherin! Slytherin values ambition, 
        cunning, leadership, and resourcefulness!" """)
        
            
    elif house == house_list[3]:
        print(""" 
        "Welcome to Ravenclaw! Ravenclaw values intelligence, 
        learning, wisdom and wit!" """)
        
    print(f"""\n {'-' * 80}""")
    
    print(f"""\t 
        {username.capitalize()}, you are from {house} in the Hogwarts 
        school of witchcraft and wizardry! Hogwarts is naming champions that 
        will be taking part in the Triwizard Tournament! Representing Hogwarts 
        are both Cedric Digory and Harry Potter! There are three tasks in the 
        Triwizard Tournament and the first two have already taken place! 
        Unfortunately, Cedric Digory was injured while flying his broom, 
        so he will not be able to participate in the third task. 
        Hogwarts is trying to find a replacement!""")

# some lines to make it more appealing to the eye

    print(f"""\n {'-' * 80}""")

    input("""<To continue press 'Enter'> \n""")

# ask the user's name so that he/she can feel like they're part of the game
    
    print("""\t
        After a lot of talk on who was going to 
        replace Cedric, it was time to choose the participant! Everyone 
        gathered around the Goblet of Fire, and waited for Professor
        Dumbledore to pick a name. """)
    
    first_stage() # user goes into the first stage of the game

def first_stage(): 
    
    # user has to choose if he's willing to participate in the tournament.
    
    print("""\t\n 
        A little note spurted out from the flames 
        and Professor Dumbledore screams out your name! 
    """)
    print(f"""\n {'-' * 80}""")
    print(f"""\t\n 
        You are shocked that your name was chosen and head down to 
        greet all the professors. Professor Dumbledore asks you, "Are you 
        ready to face the third task?"
        
        1. Yes! I am so excited to participate!
        2. No, I am scared of this tournament!
        3. Is there anyway I can give my place to someone else?
        """)

    choice = input("""
        "I know you're in shock but what is your answer?!"
    \n""").lower()

    if "1" in choice or "yes" in choice or "excited" in choice or "participate" \
    in choice:
        print(f"""\n 
        "That's what I like to hear! Ladies and Gentlemen, we were
        able to substitute Cedric Digory and will have another champion that 
        will represent Hogwarts in the third task of the Triwizard Tournament!""")
        input("""\t\tLet's go!
          
          <Press 'Enter' to continue> \n""")
        print(f""" {'-' * 80}""")  
        second_stage() # makes user go to the second stage of the game
    
    elif "2" in choice or "no" in choice or "scared" in choice or "No" in choice:
        print(f"""\n
        "What do you mean?! I'm not sure how you managed to get into
        our school with that poor attitude!"  """)
        input("""You have been asked to go back to your dorms. See you later!\n""")
        fail() # user has failed to complete the game

    elif "3" in choice or "give" in choice or "place" in choice or "someone" in\
    choice:
        print("""\n
        "No, there is not! The Goblet knows who is qualified and you
        have been chosen to represent Hogwarts! Let's get ready!"
        """)
        input("<Press 'Enter' to continue>\n")
        print(f""" {'-' * 80}""")
        second_stage() # makes user go to the second stage of the game
    
    else:
        print("""\n
        "I realize you are probably scared but I couldn't understand 
        what you tried to say! Could you please repeat your answer?" """)
        input(""""
        <Press 'Enter' to snap out of the shock and answer again!>\n """)
        first_stage() # makes user go back to the first stage of the game
        
def second_stage():
    
    # user has to answer 2 questions so Harry begins to trust user
    
    print(f"""
        The third task of the Triwizard Tournament is to go inside a
        bewithched maze and find the Goblet of Fire! 
    
        You enter the maze...
    
        Once you are inside, you lose track of where you are and suddenly Harry 
        shows up pointing his wand in your direction very agitated.""")
       
          
    print("""
        Harry goes, "Look, I think you are who you appear to be, but there are many
        ways the other competitors can appear like me, such as drinking the polly 
        juice potion! I'll have you answer these next two questions to see if you 
        really are who you say you are!" """)
    
    question_list = ['What are the names of my parents?',
                     'What are the four houses in Hogwarts?',
                     'What is the shape of the scar on my forehead?',
                     'What is the actual name of Voldemort?']
    questions = 2
    
    # while loop to answer two questions
    
    while questions > 0:
        question = random.choice(question_list) 
        input("<Press 'Enter' for your question>\n\n")
        print(question)
        print(" ")
        
        if question == question_list[0]:
            print("1) Hermione and Ron")
            print("2) James and Lily")
            print("3) Tom and Bellatrix")
            
            parents = input("> \n").lower()

            if parents == '2' or parents == 'james' or parents == 'lily'\
            or parents == 'lilly' or parents == 'james and lily':
                print("Hmmm... Good answer!\n")
                
                questions -=1

            else: 
                print(""" 
            "I knew you were an imposter!""

            Harry paralyses you by casting the spell "Petrificus Totalus!" 
            and you are eliminated from the third task!
            """)
                fail() # user loses the game
                

        
        elif question == question_list[1]:
            print("1) Gryffindor, Hufflepuff, Ravenclaw, Slytherin")
            print("2) Harry! We literally have class together! Stop with this!")
            print("3) I don't know! I am a first year!")

            four_houses = input("> \n").lower()

            if four_houses == '1' or four_houses == 'gryffindor' or four_houses \
            == 'gryf' or four_houses == 'hufflepuff' or four_houses == 'huff'\
            or four_houses == 'ravenclaw' or four_houses == 'slytherin' or \
            four_houses == 'Gryffindor, Hufflepuff, Ravenclaw, Slytherin':
                print(""" "Thanks for that. I am starting to believe in you!"
                \n""")
                
                questions -=1

            else: 
                print(""" 
                "That was not the answer I was looking for! I knew it!"
                
                Harry paralyses you by casting the spell "Petrificus Totalus!" 
                and you are eliminated from the third task!
                """)
                fail() # user loses the game

        elif question == question_list[2]:
            print("1) Magic Book")
            print("2) Wand")
            print("3) Lightning")
            
            scar = input("> \n").lower()
            
            if scar == '3' or scar == 'lightning':
                print("That's correct! Look at it!\n")
        
                questions -=1
            
            else: 
                print(""" 
                "I knew you didn't know me and wanted to take
                advantage of the situation!""
                
                Harry paralyses you by casting the spell "Petrificus Totalus!" 
                and you are eliminated from the third task!
                """)
                fail() # user loses the game
        
        elif question == question_list[3]:
            print("1) Tom Trickster")
            print("2) Tom Shelby")
            print("3) Tom Riddle")
            
            voldemort = input("> \n").lower()
            
            if voldemort == '3' or voldemort == 'riddle':
                print("Great Job! I tried to trick you with that one!\n")
                
                questions -=1
            
            else: 
                print(""" 
                "I knew you didn't know me and wanted to take
                advantage of the situation!""
                
                Harry paralyses you by casting the spell "Petrificus Totalus!" 
                and you are eliminated from the third task!
                """)
                fail() # user loses the game
     
                break
            
    print(""" "Brilliant! Now that I trust you, let's go get that goblet!" """)
    stage_three() # sends user to stage three
    
def stage_three(): # stage three begins
    print(f""" {'-' * 80}""")
    
    # user has to answer this conditional statement
    
    print("""
        You, together with Harry battle the other two champions and are
        able to reach the end of the maze, where the Goblet is located. You both 
        grab it and as soon as you touch it, the both of you end up being 
        teleported into a scary looking cemetery.
    
        Voldemort is there waiting for both of you with his scary looking face.
        He goes, "Harry Potter has come to die!" 
    
        Voldemort and Harry are both casting spells at each other and you decide
        to help! 
    
        Which spell will you use?
    
        1. Avada Kedavra!
        2. Crucio!
        3. Expelliarmus!""")
    
    choice = input(prompt = """\n 
        Harry screams, "Go on! Help me! Cast your spell!" """).lower()
    
    if '1' in choice or 'avada' in choice or 'kedavra' in choice or 'ava' \
    in choice or 'ked' in choice:
        print(""" \n
        Before you finish casting your spell, Harry stops you by
        saying, "Are you crazy?! Do you want to be sent to Azkaban!?" """)
        input("""\n
        Cast another spell! 
              
        <Press 'Enter' to try again> """)
        stage_three() # sends user back to answer stage three's question
        
    elif '2' in choice or 'crucio' in choice or 'cru' in choice or 'cio' \
    in choice:
        print(""" \n
        Before you finish casting your spell, Harry stops you by
        saying, "Are you crazy?! Do you want to be sent to Azkaban!?" """)
        input("""\nCast another spell! 
              
             <Press 'Enter' to try again> """)
        stage_three() # sends user back to answer stage three's question
        
    elif '3' in choice or 'expelliarmus' in choice or 'exp' in choice or \
    'expel' in choice:
        print("""\n
        You take out your wand, point it at Voldemort and scream,
        
        "Expelliarmus!" 
        
        The spell goes through and you disarm Voldemort leaving him without
        his wand.
        """)
        
        print(f""" {'-' * 80}""")
        
        print("""\n
        Both you and Harry take this opportunity and run for the 
        Goblet and as soon as you touch it, you are teleported back to where 
        you were supposed to be! 
        
        Both of you are greeted with loud cheers as you raise the Goblet!""")
        input("""
        <Press 'Enter' to continue> \n""")
        print(f""" {'-' * 80}""")
        champion() # user is sent to the final part of the game where he wins
    
    else:
        print("""\n"
        What are you saying?! Hurry up and cast your spell!" """)
        input("""\n
        Cast another spell! 
              
        <Press 'Enter' to try again> """)
        stage_three() # sends user back to answer stage three's question

def champion(): # user beats the game
    print("""'\n
        Congratulations! You and Harry have brought pride and honor to 
        Hogwarts by winning the Triwizard Tournament's third task! You were
        also able to defeat the Dark Lord and get out of that cemetery safely.
          """)
    print("""
        Don't think it's over, as he will return more enraged than ever!
        You have to stay ready, but for now, enjoy this remarkable achievement!""")
    
    print(f""" {'-' * 80}""")
    
    choice = input(prompt = """\n
        Would you like to play again? (yes/no)""").lower()
    
    if 'yes' in choice:
        print("""\n
        I'm glad you liked it!""")
        start()
    elif 'no' in choice:
        print("""\n
        Thank you for playing!""")
        exit(0)
    else: # relaunch the game
        print("""\n
        Please only reply with "yes" or "no." """)
        champion()
        
def fail(): # if user fails to answer the right question 
    print("""\n 
        You have failed to win the Triwizard Tournament.""")
    print("""\n
        <Game Over>""")
    
    choice = input(prompt = """\n
        Would you like to play again? (yes/no)""").lower()
    
    if 'yes' in choice:
        print("""\nI'm glad you liked it!""")
        start()
    elif 'no' in choice:
        print("""\nThank you for playing!""")
        exit(0)
    else: # relaunch the game
        print("""\nPlease only reply with "yes" or "no." """)
    fail()
    
start()

```
